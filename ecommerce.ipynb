```python
#Rating Distribution:

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\kirut\Downloads\Ecommerce_Delivery_Analytics_New.csv")
sns.countplot(x='Service Rating', data=df)
plt.title('Service Rating Distribution')
plt.show()
```

```python
#Refund analysis

df['Refund Requested'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Refund Requested')
plt.show()
```

```python
# Delivery Delay Analysis:

sns.boxplot(x='Delivery Delay', y='Delivery Time (Minutes)', data=df)
plt.title('Delivery Delay vs Delivery Time')
plt.show()
```

```python
#Platform wise Rating:
df.groupby('Platform')['Service Rating'].mean().plot(kind='bar', color='steelblue')
plt.title('Avg Rating by Platform')
plt.show()
```
